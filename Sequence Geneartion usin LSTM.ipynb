```python
import tensorflow as tf
import numpy as np
```

```python
data=[i for i in range(0,10000) if i%2==0]
```

```python
data
```

```python
length=20+1
lines=[]
for i in range(length,5000):
    num=list(data[i-length:i])
    lines.append(num)
```

```python
lines=np.array(lines)
```

```python
lines
```

```python
X,y=lines[:,:-1],lines[:,-1]
```

```python
X.shape
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
y=to_categorical(y,num_classes=10000)
```

```python
seq_length=X.shape[1]
```

```python
model =Sequential()
model.add(Embedding(10000,20,input_length=seq_length))
model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100))
model.add(Dense(y.shape[1],activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X,y,batch_size=256,epochs=300)
```

```python
def generate_text_seq(model,test_data,n_words):
    text=[]
    for _ in range(n_words):
        encoded=pad_sequences([test_data],20,truncating='pre')
        
        y_predict=model.predict_classes(encoded)
        test_data.append(y_predict[0])
        text.append(y_predict[0])
    return text
```

```python
generate_text_seq(model,test_data,10)
```
